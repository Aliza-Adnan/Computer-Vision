```python
import numpy as np
import matplotlib.pyplot as plt
fouriert=np.fft.fft2(img)
ji=np.fft.fftshift(fouriert)
f=np.log(abs(ji))
plt.imshow(f)
u=ji
k=ji
o=ji
```

```python
x=f.shape[0]/2
y=f.shape[1]/2
thresh=40

for i in range(f.shape[0]):
  for j in range(f.shape[1]):
    dist=np.sqrt(((x-i)**2)+((y-j)**2))
    if dist<thresh:
      u[i,j]=0


```

```python
s=np.real(np.fft.ifft2(np.fft.ifftshift(u)))
plt.imshow(s)
```

```python
res=np.log(abs(u))
plt.imshow(res)
```
