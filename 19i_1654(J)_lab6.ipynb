```python
l= cv2.imread("/content/drive/MyDrive/brad.jpg", -1)

l=cv2.resize(l,(500,500))
image4= cv2.rotate(l, cv2.ROTATE_180)
cv2_imshow(image4)
```

```python
image2 = cv2.rotate(l, cv2.ROTATE_90_COUNTERCLOCKWISE)
cv2_imshow(image2)
```

```python
sift = cv2.xfeatures2d.SIFT_create()
keypoints_2, descriptors_2 = sift.detectAndCompute(image4,None)

image4 = cv2.drawKeypoints(image4,keypoints_2,image4)

cv2_imshow(image4)
```

```python
sift = cv2.xfeatures2d.SIFT_create()
keypoints_3, descriptors_3 = sift.detectAndCompute(image2,None)

image2 = cv2.drawKeypoints(image2,keypoints_3,image2)

cv2_imshow(image2)
```

```python
bf = cv2.BFMatcher(cv2.NORM_L1, crossCheck=True)

matches = bf.match(descriptors_3,descriptors_2)
matches = sorted(matches, key = lambda x:x.distance)

img3 = cv2.drawMatches(image2, keypoints_3, image4, keypoints_2, matches[:50],None)
plt.imshow(img3),plt.show()
```

```python
cv2_imshow(img3)
```
